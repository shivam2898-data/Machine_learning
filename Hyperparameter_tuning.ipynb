```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"D:\Python\Data set\covid_toy.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer()
```

```python
df['fever']=si.fit_transform(df[['fever']])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
df['cough']=lb.fit_transform(df['gender'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
df.head()
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score  # Accuracy_score--> for categorigal data
```

```python
accuracy_score(y_test ,y_pred)
```

# Hyperparameter Tuning

```python
param_grid=[
    {'penalty':['l1','l2','elasticnet','none'],
    'C':np.logspace(-4,4,20),
    'solver':['lbfgs','newton-sg','liblinear','sag','saga'],
    'max_iter':[100,1000,2500,5000]}
]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(lr,param_grid=param_grid,cv=3,verbose=True,n_jobs=-1)
```

```python
best_clf=clf.fit(x_train,y_train)
```

```python
best_clf.best_estimator_
```

```python
print(f'Accuracy-:{best_clf.score(x_train,y_train):.3f}')
```

```python
df=pd.read_csv(r"D:\Python\Data set\titanic.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
si=SimpleImputer()
```

```python
df['Age']=si.fit_transform(df[['Age']])
df['Fare']=si.fit_transform(df[['Fare']])
```

```python
df=df.drop(columns=['Cabin'])
df=df.drop(columns=['Ticket'])
```

```python
df.isnull().sum()
```

```python
lb=LabelEncoder()
```

```python
df['Name']=lb.fit_transform(df['Name'])
df['Sex']=lb.fit_transform(df['Sex'])
df['Embarked']=lb.fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
x=df.drop(columns=['Survived'])
y=df['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
accuracy_score(y_test ,y_pred)
```

```python
param_grid=[
    {'penalty':['l1','l2','elasticnet','none'],
    'C':np.logspace(-4,4,20),
    'solver':['lbfgs','newton-sg','liblinear','sag','saga'],
    'max_iter':[100,1000,2500,5000]}
]
```

```python
clf=GridSearchCV(lr,param_grid=param_grid,cv=3,verbose=True,n_jobs=-1)
```

```python
best_clf=clf.fit(x_train,y_train)
```

```python
best_clf.best_estimator_
```

```python
print(f'Accuracy-:{best_clf.score(x_train,y_train):.3f}')
```
