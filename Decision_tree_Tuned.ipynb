# Decision Tree

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"D:\Python\Data set\covid_toy.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer()
```

```python
df['fever']=si.fit_transform(df[['fever']])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
df['cough']=lb.fit_transform(df['cough'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
x=df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(x_train , y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# create the parameter grid based on the results of random search
params={
    'max_depth':[2,3,5,10,20],   # numbers of leaf node required
    'min_samples_leaf':[5,10,20,50,100], #Minimum number of child node leaf required to be in a leaf node.
    'criterion':['gini','entropy']
```

```python
grid_search=GridSearchCV(estimator=dt,    #cv-cross validation-->divide in folder
                        param_grid=params, #n_jobs=-1---> CPU
                        cv=6, n_jobs=-1, verbose=1, scoring='accuracy')
```

```python
grid_search.fit(x_train ,y_train)
```

```python
grid_search.best_estimator_
```

```python
y_pred_2=grid_search.predict(x_test)
```

```python
accuracy_score(y_test ,y_pred_2)
```
